```python
# Функция для преобразования строки в camel case
def to_camel_case(s):
    # Разбиваем строку на слова по '_'
    words = s.split('_')
    # Превращаем первое слово в нижний регистр, остальные - в верхний регистр
    return words[0].lower() + ''.join(word.capitalize() for word in words[1:])

# Применяем функцию ко всем столбцам DataFrame
seller_data.columns = [to_camel_case(col) for col in seller_data.columns]

# Выводим обновленный DataFrame для проверки
seller_data.head()
```

## Задача 1

На основе исходных данных и данных из таблицы рассчитать:
   - средний чек с каждой продажи

## Задача 2

Опираясь на расходы, исходные данные, данные из таблицы и те данные, которые мы рассчитали, провести расчёты ключевых
показателей юнит-экономики для покупателей и продавцов (CAC, LTV и ROI), определить проблемные зоны в модели бизнеса и
предложить планы по оптимизации CAC и увеличению LTV.

```python
# Исходные данные
subscription_cost = 3000  # стоимость подписки в месяц
avg_order_value = 2009.67  # средний чек с продажи
avg_commission = 0.1  # процент средней комиссии с продажи
avg_seller_income = 3338.40  # средний доход продавца
marketing_expenses = 3000000  # маркетинговые расходы в месяц
avg_buyers_per_month = 2000  # средний объем привлекаемых покупателей в месяц
avg_sellers_per_month = 500  # средний объем привлекаемых продавцов в месяц
operational_expenses = 1500000  # операционные расходы в месяц
seller_service_cost = 1000  # расходы на обслуживание каждого продавца в месяц
buyer_retention_rate = 0.20  # удержание покупателей
seller_retention_rate = 0.40  # удержание продавцов
average_interaction_months_buyer = 24  # средняя продолжительность взаимодействия для покупателей (в месяцах)
average_interaction_months_seller = 18  # средняя продолжительность взаимодействия для продавцов (в месяцах)

# Расчет CAC для покупателей
CAC_buyer = marketing_expenses / avg_buyers_per_month

# Расчет CAC для продавцов
CAC_seller = marketing_expenses / avg_sellers_per_month

# LTV для покупателей (с учетом комиссии)
LTV_buyer = (avg_order_value * avg_commission)* average_interaction_months_buyer* buyer_retention_rate

# LTV для продавцов (с учетом комиссии)
LTV_seller = (avg_seller_income - seller_service_cost)* average_interaction_months_seller * seller_retention_rate

# ROI для покупателей (с учетом комиссии)
ROI_buyer = (LTV_buyer-CAC_buyer) / CAC_buyer

# ROI для продавцов
ROI_seller = (LTV_seller-CAC_seller) / CAC_seller

# Форматирование и вывод результатов
results = {
    "Показатель": ["CAC (покупатели)", "LTV (покупатели)", "ROI (покупатели)",
                   "CAC (продавцы)", "LTV (продавцы)", "ROI (продавцы)"],
    "Значение": [CAC_buyer, LTV_buyer, ROI_buyer, CAC_seller, LTV_seller, ROI_seller]
}

results_df = pd.DataFrame(results)

# Вывод результатов
print(results_df)
```

## Выводы

1. Покупатели:
   - CAC (покупатели): 1500
   - LTV (покупатели): 964.64
   - ROI (покупатели): -0.36

   Значения показывают, что затраты на привлечение покупателей (CAC) превышают доход, который они приносят (LTV). Это приводит к отрицательному ROI, что указывает на убыточность бизнеса в сегменте покупателей.
2. Продавцы:
   - CAC (продавцы): 6000
   - LTV (продавцы): 16836.48
   - ROI (продавцы): 1.81

   В этом сегменте затраты на привлечение продавцов (CAC) значительно ниже их LTV, что приводит к положительному ROI и указывает на прибыльность бизнеса в сегменте продавцов.





## Проблемные зоны

1. Отрицательный ROI по покупателям: Это основная проблема. Бизнес теряет деньги на привлечении покупателей, что делает модель неустойчивой.
2. Высокий CAC по покупателям: Затраты на привлечение покупателей слишком высоки по сравнению с их ценностью.
3. Неоптимизированный процесс привлечения покупателей: Возможно, маркетинговые стратегии неэффективны или неправильные.

## Рекомендации по оптимизации CAC и увеличению LTV

1. Оптимизация маркетинговых расходов:
   - Провести анализ текущих каналов привлечения покупателей и определить, какие из них наиболее эффективны.
   - Уменьшить затраты на менее эффективные каналы и перераспределить бюджет на те, которые показывают лучшие результаты.
2. Улучшение предложения для покупателей:
   - Провести опросы и исследования, чтобы понять потребности и предпочтения покупателей. Это поможет создать более привлекательное предложение.
   - Рассмотреть возможность внедрения программ лояльности или скидок, чтобы увеличить удержание клиентов и, соответственно, LTV.
3. Увеличение среднего чека:
   - Разработать стратегии кросс-продаж и апсейлов для увеличения среднего чека от покупателя.
   - Рассмотреть возможность расширения ассортимента товаров или услуг, чтобы увеличить ценность для клиентов.
4. Оптимизация воронки продаж:
   - Улучшить конверсию на всех этапах воронки, чтобы больше потенциальных клиентов становились покупателями. Это может включать улучшение сайта, оптимизацию процесса оформления заказа и повышение качества обслуживания клиентов.
5. Изучение альтернативных моделей монетизации:

   - Инвестировать в обучение команды по обслуживанию клиентов, чтобы повысить удовлетворенность и удержание покупателей.



## Задача 3

Когортный анализ для выявления спада активности покупателей и продавцов

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Создание когорты по месяцу регистрации
cohort_analysis_data['cohort'] = cohort_analysis_data['registrationdate'].dt.to_period('M')

# Подсчет количества пользователей в каждой когорте
cohort_counts = cohort_analysis_data.groupby('cohort').size().reset_index(name='total_users')

# Подсчет активных пользователей по месяцам
activity_counts = cohort_analysis_data.groupby('cohort').agg({
    'activity1month': 'sum',
    'activity3month': 'sum',
    'activity6month': 'sum',
    'activity12month': 'sum'
}).reset_index()

# Объединение данных о когорте и активности
cohort_activity = pd.merge(cohort_counts, activity_counts, on='cohort')

# Расчет процентов удержания
cohort_activity['activity1month'] = cohort_activity['activity1month'] / cohort_activity['total_users'] * 100
cohort_activity['activity3month'] = cohort_activity['activity3month'] / cohort_activity['total_users'] * 100
cohort_activity['activity6month'] = cohort_activity['activity6month'] / cohort_activity['total_users'] * 100
cohort_activity['activity12month'] = cohort_activity['activity12month'] / cohort_activity['total_users'] * 100

# Создание таблицы для тепловой карты
retention_matrix = cohort_activity.set_index('cohort')[['activity1month', 'activity3month', 'activity6month', 'activity12month']]

# Визуализация тепловой карты
plt.figure(figsize=(10, 6))
sns.heatmap(retention_matrix, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Удержание (%)'})
plt.title('Когортный анализ: Удержание покупателей по месяцам')
plt.xlabel('Период (месяцы)')
plt.ylabel('Когорты (месяц регистрации)')
plt.xticks(ticks=[0.5, 1.5, 2.5, 3.5], labels=['1 месяц', '3 месяца', '6 месяцев', '12 месяцев'], rotation=45)
plt.yticks(rotation=0)
plt.show()
```

Основное целевое действие для них — продажа товаров, и именно их активность определяет успешность взаимодействия с платформой. Однако спустя три месяца удержание продавцов резко падает, что указывает на проблемы с долгосрочным удержанием;

•	сезонные изменения. Первые месяцы работы продавцов на платформе показывают хорошие результаты. Это говорит о привлекательности платформы и удобстве интерфейса. Но спустя три месяца наблюдается резкое снижение активности, что может быть связано с недостаточной ценностью платформы на долгосрочной основе.


## Выводы и рекомендации

•	оптимизация стратегии удержания. Для повышения уровня удержания необходимо проанализировать, какие маркетинговые кампании и стратегии были использованы в успешные периоды и можно ли их масштабировать на другие месяцы;

•	фокус на качестве обслуживания. Особое внимание стоит уделить качеству обслуживания и интерфейсу, чтобы минимизировать неудовлетворенность пользователей и продавцов, которая может быть причиной их ухода с платформы;

•	снижение операционных расходов. Операционные расходы на продавцов могут быть снижены за счет автоматизации процессов и использования чат-ботов для поддержки, что также позволит улучшить взаимодействие с платформой.


## Проблемы, которые требуют решения

•	высокие затраты на привлечение клиентов (CAC) при низком уровне удержания;

•	низкое удержание пользователей, что приводит к сокращению их жизненного цикла на платформе;

•	ограниченные возможности увеличения LTV (пожизненной ценности клиента) как для покупателей, так и для продавцов.


## Решения

•	сегментация клиентов для выявления наиболее перспективных групп;

•	оптимизация каналов привлечения пользователей и проведение маркетинговых экспериментов для улучшения удержания;

•	введение программ лояльности и улучшение интерфейса для удобства пользователей и продавцов.
